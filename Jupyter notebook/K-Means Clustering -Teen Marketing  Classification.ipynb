### Outliers imputation

#### For Age

```python
#for value in colname:
q1 = df['age'].quantile(0.25) #first quartile value
q3 = df['age'].quantile(0.75) # third quartile value
iqr = q3-q1 #Interquartile range
low  = q1-1.5*iqr #acceptable range
high = q3+1.5*iqr #acceptable range
```

```python
df_include = df.loc[(df['age'] >= low) & \
                    (df['age'] <= high)]
df_exclude = df.loc[(df['age'] < low) | (df['age'] > high)]
```

```python
print(df_include.shape)
print(df_exclude.shape)
```

```python
print(low)
```

```python
age_mean=int(df_include.age.mean()) #finding the mean of the acceptable range
print(age_mean)
```

```python
df_exclude.age=age_mean
```

```python
#getting back the original shape of df
df_rev=pd.concat([df_include,df_exclude],axis=0) #concatenating both dfs to get 
#the original shape
df_rev.shape
```

```python
df_rev.boxplot(column="age")
```

```python
plt.scatter(data=df_rev, y='friends',x='age')
plt.xlabel("Age")
plt.ylabel("Friends")
plt.show()
```

### Label Encoding

```python
# For preprocessing the data 
from sklearn import preprocessing 
le=preprocessing.LabelEncoder()

col=['gradyear','gender']
for i in col:
    df_rev[i]=le.fit_transform(df_rev[i])
    
    le_name_mapping=dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature',i)
    print('le_name_mapping',le_name_mapping)
    print()
```

```python
# convert datafram type to array because array is lighter than dataframe by using .values():
X=df_rev.iloc[:,:]
X.head()
```

### KMeans Clustering Model

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
```

```python
wsse = []
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, random_state = 10) 
    kmeans.fit(X) 
    wsse.append(kmeans.inertia_)  
plt.plot(range(1, 11), wsse) 
# plt.scatter(range(1, 11),wsse)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WSSE')
# plt.grid()
plt.show()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Fitting K-Means to the dataset:
kmeans=KMeans(n_clusters=5, random_state=10)
# Y_pred=kmeans.fit_predict(X)
kmeans.fit(X)
Y_pred1=kmeans.predict(X)
```

```python
Y_pred1
```

```python
# to see how many iteration it perform:
kmeans.n_iter_
```

## K-Means using PCA

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
```

```python
# Applying PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = None) #to view the entire eigen vector,no subsetting
X = pca.fit_transform(X)
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

```python
# Applying PCA
from sklearn.decomposition import PCA
pca = PCA(n_components =2) #to view the entire eigen vector,no subsetting
X= pca.fit_transform(X)
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

```python
# Fitting K-Means to the dataset:
kmeans_pca=KMeans(n_clusters=5, random_state=10)
# Y_pred=kmeans.fit_predict(X)
kmeans_pca.fit(X)
Y_pred=kmeans_pca.predict(X)
```

```python
Y_pred
```

```python
# to see how many iteration it perform:
kmeans_pca.n_iter_
```

```python
df_rev['Clusters']=Y_pred1
df_rev.head(10)
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X, Y
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, kmeans_pca.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.3, cmap = ListedColormap(('red', 'green', 'blue','orange','yellow')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue','orange','yellow'))(i), label = j)
plt.title('SVM (Training set)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```
